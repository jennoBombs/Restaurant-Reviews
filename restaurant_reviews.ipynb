```python
import pandas as pd
```

```python
# read the business.json into a dataframe
business_df = pd.read_json('business.json',lines=True)
business_df.head(20)
```

```python
#R3 - Which state has the most businesses listed? Count the occurrances & return the index
biz_count = business_df['state'].value_counts()
max_state_business = biz_count.idxmax()
print(max_state_business)
```

```python
#initialize an empty list
chunks = []
total_count = 0

#iterate over the yelp JSON file
for chunk in pd.read_json('yelp_dataset_review_data.json', lines=True, chunksize = 500000):
    chunks.append(chunk)
    
#concat the chunks into a single df
yelp_df = pd.concat(chunks, ignore_index=True)
yelp_df.head()
```

```python
zip_filter_df = business_df[business_df['postal_code'] == '85086']

restaurant_df = zip_filter_df[zip_filter_df['categories'].notnull() & zip_filter_df['categories'].str.contains('restaurant', case=False)]
top_restaurants = restaurant_df.sort_values(by='stars', ascending=False).head(5)
print(top_restaurants)
```
